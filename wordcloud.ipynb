```python
"""
Masked wordcloud
================
Using a mask you can generate wordclouds in arbitrary shapes.
"""

from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import sys
import pandas as pd

from wordcloud import WordCloud, STOPWORDS
```

```python
#d = path.dirname(sys.argv[0])

dict1 = {}
list1 = ['the','of','a','was','were','are','we','you','at','this',
         'he','and','to','not','their','have','will','has','for','be','his','would',
        'which','that','an','one','with','as','is','been','or','I','your','they',
        'in','on','The','from','it','by','who','but','had','its','up','more','about',
         'can','than','all','first','said','also','after','just','our','into','what',
        'other','when','two','there','but','even']

# Read the whole text.
with open('C:\Users\hanqing\Desktop\part-22') as f:
    for line in f:
        line = line.strip('\n')

        line = line.split('\t')
        word, count = line
        if word not in list1:
            dict1[word] = float(count)
```

```python
dict1
```

```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open('C:\Users\hanqing\\Desktop\\ren.png'))

wordcloud1 = WordCloud(width = 800, height = 400, background_color="white", max_words=2000, mask=alice_mask, 
                       font_step = 1, min_font_size = 3, scale = 5)
#text = {key:value for key,value in dict1.iteritems()}
#text = {'hello':5}
#print(text)
wordcloud1.fit_words(dict1)


# # generate word cloud
# wc.generate()

# store to file
wordcloud1.to_file("C:\Users\hanqing\\Desktop\\alice.png")

# show
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis("off")
plt.figure()
#plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
#plt.axis("off")
plt.show()
```
